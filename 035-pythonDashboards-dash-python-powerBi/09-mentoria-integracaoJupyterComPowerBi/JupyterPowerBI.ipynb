### Agora podemos interagir com o relatório

```python
# agora podemos utilizar o relatorio armazenado na variavel relatorio conforme for interessante
# vamos pegar as paginas do relatorio:
paginas=relatorio.get_pages()

print('Número de paginas do relatório: '+str(len(paginas)))
```

- Como assim? Vamos entender

```python
for pagina in paginas:
    print('*'*10)
    print(pagina)
    if pagina['displayName']=='Relatório de Vendas':
        pagina_relatorio_vendas=pagina

# observe na resposta que temos 13 paginas, mas apenas 3 estão visiveis (visibility: 0)
# os tooltips e os relatorios sao considerados páginas
```

### Vamos puxar agora 1 página: Relatório de Vendas 

```python
print(pagina_relatorio_vendas)

# o metodo visuals_on_page() que so funciona no relatorio, devolve os graficos que existem na pagina e todos os outros tipos de 'visuais', a referencia para usa-lo deve ser sempre o campo nome
lista_graficos=relatorio.visuals_on_page(pagina_relatorio_vendas['name'])
print(len(lista_graficos))

for grafico in lista_graficos:
    print('*'*10)
    print(grafico)
    if grafico['type']=='areaChart':
        grafico_area=grafico

# observe que no par chave, valor 'type':'tipo' tipo será o tipo de grafico (area, barra, mapa, etc)
```

### Pegar os gráficos de área (ou gráfico com determinada condição, como Título) 

```python
print(grafico_area)
```

### Exportar informações do Power BI para o Python 

```python
# as referencias a serem utilizadas devem ser o nome da pagina seguida do nome do grafico, ambos estao referenciados nesse caso pela chave ['name'] do dicionario de cada um de seus elementos
# vamos colocar rows=100 pois o numero de linhas pode ser maior, se for menor nao pegara todos os dados, o padrao caso o valor de rows nao seja mencionado será de 10 linhas
# a pagina a ser capturada deve estar como a pagina atual no power bi
infos_grafico=relatorio.export_visual_data(pagina_relatorio_vendas['name'],grafico_area['name'],rows=100)

print(infos_grafico)

# observe que ele retornara as 12 linhas do grafico mesmo com rows=100
```

```python
# quanto tiver um texto que tem uma tabela e quiser que o panda leia essa tabela basta colocar esse texto dentro de um StringIO

import pandas as pd
from io import StringIO

tabela=pd.read_csv(StringIO(infos_grafico))
display(tabela)
```

### Outra forma de exportar 

```python
# o export_data_type=1 significa que ele vai ignorar os filtros do grafico e trazer todas as informacoes e medidas presentes no power bi
infos_grafico2=relatorio.export_visual_data(pagina_relatorio_vendas['name'],grafico_area['name'],rows=100, export_data_type=1)

print(infos_grafico2)

tabela2=pd.read_csv(StringIO(infos_grafico2))
display(tabela2)
```

### Trabalhar com Filtros 

- Identificar se tem filtro aplicado

```python
# vai imprimir os filtros que existem no relatorio em forma de dicionario (par chave valor)
# no caso serao os filtros que estao na aba da direita do power bi (padrao do relatorio), nao os filtros colocados dentro do dashboard em si, os filtros usados para construir o power bi
print(relatorio.get_filters())

# está filtrando a tabela CadastroLojas, coluna Loja
```

```python
# criando filtro para os anos, importante salientar que o filtro colocado só será valido para o dashboard dentro do jupyter, nao sera exportado para o dashboard que está no power bi
def filtrar_relatorio(anos):
    filtro_anos={
        '$schema': 'http://powerbi.com/product/schema#basic',
        'target': {'table': 'CadastroLojas', 'column': 'Ano'},
        'filterType': 1,
        'operator': 'In',
        'values': [anos],
        'requireSingleSelection': False
    }
    relatorio.remove_filters()
    relatorio.update_filters([filtro_anos])
```

```python
from ipywidgets import interact

interact(filtrar_relatorio,anos=[2016,2017,2018])
```
